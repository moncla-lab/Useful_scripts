```python
import re
import os
import pandas as pd
from Bio import SeqIO
from datetime import datetime
```

```python
# This code generates a list of strains given a FASTA or TSV
```

```python
#Function: Read in the FASTA as a dataframe

def fasta_reader(path_to_fasta, output_name):
    fasta_data = []
    FASTANAME = path_to_fasta
        
    with open(FASTANAME) as f:
        header = ""
        sequence = ""
        for line in f:
            if line.startswith(">"):
                if header != "":
                    fasta_data.append({"header": header, "sequence": sequence})
                header = line.strip() 
                sequence = ""
            else:
                sequence += line.strip()
        fasta_data.append({"header": header, "sequence": sequence}) #last line 
        
    globals()[output_name] = pd.DataFrame(fasta_data)

    return
```

```python
# Read in the FASTA or TSV

FILEPATH = "../../H5-genotypes/BEAST/empirical_trees/submission_files/2025-10-21_h5_emptree_A2_constant_MLE/inputs/A2_subsampled_genome.fasta"

#tsv_data = pd.read_csv(FILEPATH, sep = '\t')

fasta_reader(FILEPATH, "fasta_data")
```

```python
# Function: remove the carrot from the header

def header_to_strain(df):
    df["strain"] = df["header"].str.replace(">","")

    return(df)
```

```python
# If the format is FASTA, make a strain column

fasta_data_strain = header_to_strain(fasta_data)
```

```python
# Function: turn the "strain" column into a list

def strain_to_list(df):

    strain_list = df["strain"].to_list()

    return(strain_list)
```

```python
strain_list = strain_to_list(fasta_data_strain)
```

```python
OUTPUT_PATH = "../../H5-genotypes/BEAST/empirical_trees/submission_files/2025-10-21_h5_emptree_A2_constant_MLE/inputs/A2_strains.txt"

with open(OUTPUT_PATH, 'w') as outfile:
  outfile.write('\n'.join(str(i) for i in strain_list))
```
